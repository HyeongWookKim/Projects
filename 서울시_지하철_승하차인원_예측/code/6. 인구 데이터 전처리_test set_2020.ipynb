        - 30대

        - 40대
    - pop_50
        - 50대
    - pop_60
        - 60대

# 조인키 생성
- 지하철 데이터 셋을 기준으로 인구 데이터 셋과 기상 데이터 셋을 결합시켜 줄 것이므로, 이를 위한 조인키가 필요하다.
- 인구 데이터의 'dong' 변수의 경우, "종로구_청운효자동" 이런 식으로 자치구와 행정동이 같이 입력되어 있다.
- 따라서 다음과 같이 조인키를 생성해주겠다.
    1. 앞서 했던 방법대로 먼저 문자열을 결합해준다.
        - ex) "2018-01-01_종로구_청운효자동"
    2. 만들어진 조인키 변수에서 자치구에 해당 되는 부분은 제거해준다.
        - ex) "2018-01-01_청운효자동"

````python
# datetime to string
pop_total_final['date'] = pop_total_final['date'].astype(str)

# 문자열 결합
pop_total_final['pop_join_key'] = pop_total_final['date'] + '_' + pop_total_final['dong']
# '_'를 기준으로 문자열을 잘라서 리스트에 담기
pop_join_key_list = pop_total_final['pop_join_key'].str.split('_').to_list()
````

````python
# '자치구' 부분을 제외해 준 뒤에 문자열 재결합
result = []
for i in tqdm(pop_join_key_list):
    result.append('_'.join([i[0], i[2]]))

# 지하철 데이터와 결합할 때 사용할 조인키
pop_total_final['pop_join_key'] = result
pop_total_final['pop_join_key']
````
