## 데이터 셋의 형태 변경
- 현재 주어진 데이터 셋의 형태를 보면, 시간대별 승/하차 인원 변수(0, 5 ~ 23)가 컬럼으로 들어가있다.
- 때문에 시간대별 승/하차 인원 변수(0, 5 ~ 23)를 시간대(hour) 변수와 승/하차 인원수(boarding_num) 변수로 분리되도록 데이터 셋의 형태를 변환해주도록 하겠다.
    - 즉, **columns to rows** 변환

```python
# '0, 5~23' 변수들이 문자열로 입력되어 있다는 점을 고려해서 리스트에 담아주기
pivot_values = []
for i in list(range(5, 24)) + [0]:
    pivot_values.append(i)
    
# 리스트에 담긴 숫자들을 오름차순 정렬
pivot_values = sorted(pivot_values)
pivot_values
```

```python
# DataFrame to Pivot Table
subway_pivot = pd.pivot_table(subway_boarding_total, index = ['date', 'line', 'station_num', 'station_name', 'boarding'], 
                              values = pivot_values)
subway_pivot
```

```python
# stack 함수를 통한 변환 후, DataFrame으로 변경
subway_boarding_total2 = pd.DataFrame(subway_pivot.stack()).reset_index()
subway_boarding_total2
```

- stack 함수를 통해 만들어진 DataFrame의 level_5 변수와 0 변수의 변수명을 변경해주겠다.
    - level_5 = hour
    - 0 = boarding_num
- 그리고 만들어진 데이터 셋을 date와 hour 변수를 기준으로 오름차순 정렬해주도록 하겠다.

```python
subway_boarding_total2.rename(columns = {'level_5': 'hour', 0: 'boarding_num'}, inplace = True)
subway_boarding_total2.sort_values(by = ['date', 'hour'], axis = 0, inplace = True)
subway_boarding_total2
```

## 4개의 데이터 셋 최종 결합

```python
dfs = [subway_area, subway_address, subway_type, subway_info]
subway_information_total = reduce(lambda df_left, df_right: pd.merge(df_left, df_right, how = 'inner', on = ['station_name', 'line']), dfs)
subway_information_total
```

# 조인키 생성
- 지하철 데이터 셋을 기준으로 인구 데이터 셋과 기상 데이터 셋을 결합시켜 줄 것이므로, 이를 위한 조인키가 필요하다.
- 따라서 지하철 데이터 셋과 인구 데이터 셋을 결합하기 위한 조인키 1개, 그리고 지하철 데이터 셋과 기상 데이터 셋을 결합하기 위한 조인키 1개를 만들어주도록 하겠다.
    1. 지하철 데이터 셋과 인구 데이터 셋을 결합하기 위한 조인키(pop_join_key)
        - ex) "2018-01-01_회현동"
    2. 지하철 데이터 셋과 기상 데이터 셋을 결합하기 위한 조인키(weather_join_key)
        - ex) "2018-01-01_중구"

```python
# datetime to string
subway_final['date'] = subway_final['date'].astype(str)

# 인구 데이터와 결합할 때 사용할 조인키
subway_final['pop_join_key'] = subway_final['date'] + '_' + subway_final['dong']
# 기상 데이터와 결합할 때 사용할 조인키
subway_final['weather_join_key'] = subway_final['date'] + '_' + subway_final['gu']
```
